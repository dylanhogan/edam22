### Set up environment
In this course we will set-up our R environment through `conda` enviornment. With conda, you can create, export, list, remove, and update environments that have different versions of R and/or packages installed in them. 

**What is a conda environment?** A conda environment is a directory that contains a specific collection of conda packages that you have installed. For example, you may have one environment with NumPy 1.7 and its dependencies, and another environment with NumPy 1.6 for legacy testing. If you change one environment, your other environments are not affected. See more explanation <a href="https://docs.conda.io/projects/conda/en/latest/user-guide/concepts/environments.html" target="_blank">here</a>.

**Why are we using conda environment instead of directly install packages on my computer?** So you can easily switch environments and "protect" one environment from others. Open source programming languages like Python or R are free to use but also comes with no warranties. Some packages may not work with older versions of R, and some packages may not work with each other. By using an environment, you conceal the R of a particular version together with the packages into an environment. You can create multiple environments for different courses. If you "screwed-up" your environment by installing conflicting packages, you may simple delete that environment and start clean again. 

#### 1. Install Anaconda
1. We suggest to use the `conda` enviornment with Jupyter Notebook for homework and final exam, you need to install Anaconda first following <a href="https://www.anaconda.com/products/individual" target="_blank">here</a>.
2. Make sure to add Anaconda to your system environment during the installation so you can use `conda` command from <span style="color: blue;">Anaconda Prompt</span> / <span style="color: red;">Terminal</span>.

#### 2. Add conda-forge to Anaconda channels
Channels are locations where Navigator and conda look for packages. The R environment in Anaconda defaults channel does not support current version of R and some package we will use in this course.
1. Open <span style="color: blue;">Anaconda Prompt</span> / <span style="color: red;">Terminal</span>
2. Enter `conda config --add channels conda-forge`
3. Enter `conda config --set channel_priority strict`

#### 3. Create R environment in conda with Jupyter and essiential packages
1. Open <span style="color: blue;">Anaconda Prompt</span> / <span style="color: red;">Terminal</span>
2. Enter `conda create --name myenv r-essentials r-base` (Note: replace `myenv` with your enviroment name)
3. Enter `y` when required

#### 4. Connect Jupyter with R
The last step is to set-up `IRkernel` let Jupyter use R and packages from the installed environment.
1. Open <span style="color: blue;">Anaconda Prompt</span> / <span style="color: red;">Terminal</span>
2. Enter `conda activate myenv` (Note: replace `myenv` with your enviroment name) to activate the environment
3. Enter `conda install -c conda-forge r-irkernel` to install  IRkernel, then enter `R -e 'IRkernel::installspec()'` to set it up in R.

#### 5. Run Jupyter Notebook with R
1. Download file `set_up_Jupyter+R.ipynb` from coursework
2. Open <span style="color: blue;">Anaconda Prompt</span> / <span style="color: red;">Terminal</span>
3. Enter `conda activate myenv` (Note: replace `myenv` with your enviroment name)
4. Enter `jupyter notebook`
5. In Files under Jupyter Notebook, locate the downloaded file
6. Open the file and you should see the R logo on the top-right corner
7. Select the cell below and click Run. Hotkey: `esc` (in case you are in the editing mode), then <span style="color: blue;">ctrl + enter<span> / <span style="color: red;">control + return</span>

### Install Packages

We fisrt practice installing packges from conda/conda-forge. We will install the `ks` package.


1. Open <span style="color: blue;">Anaconda Prompt</span> / <span style="color: red;">Terminal</span>
2. Enter `conda activate myenv` (Note: replace `myenv` with your enviroment name) **Always remember to activate your environment before installing packages or starting Jupyter Notebook, you can always check which environment you are at now by looking at the beginning of your terminal command line**
3. Enter `conda search r-ks` **We first search the packages and check if it's available in conda/conda-forge. When using conda to install R packages, you will need to add r- before the regular package name.**
4. Since it is available, enter `conda install r-ks`
5. You can close the terminal or move to the Step 3 in the next section.

For more information, please find the <a href="https://docs.conda.io/projects/conda/en/latest/user-guide/tasks/manage-pkgs.html#id4" target="_blank">Managing Packages User Guide</a>.

```r
!ls
```

We now practice installing non-conda packages and using them in R. **The differences between non-conda packages and conda packages cause certain unavoidable limits in compatibility. Always try to install packages in conda first. We will note if non-conda packages need to be installed in homework and exams**

We will install the `waterData` dataset, and do a sample plot using data from it.
1. Open <span style="color: blue;">Anaconda Prompt</span> / <span style="color: red;">Terminal</span>
2. Enter `conda activate myenv`
3. Enter `R` to open R console
4. Inside R console, enter `install.packages('waterData')`.
    1. **<span style="color: red;">Important for Mac</span>:** If you are experiencing errors during the installation such as the terminal showing `'stdlib.h' file not found`, you need to install and/or reset your gcc compiler. 
    2. To do this, exit R console or open a new terminal, enter (whether you are inside your environment or not does not matter for this set up) `xcode-select --install` to install xcode (this may take a while), which will install the gcc compiler for Mac. 
    3. After installation completed or it shows you already installed xcode, type `xcode-select --reset` to reset.
    4. Now try to install the waterData package again and it should not return any more errors.
5. After installed the waterData package, quit R and start Jupyter Notebook (or restart Kernel if you are already in Jupyter).
6. Run the following code. The package `tidyverse` should have already been installed in your R environment, if not, repeat the above step to install `tidyverse`. Ignore conflicts messages.

```r
library(waterData)
library(tidyverse)

sflow <- importDVs("05054000", sdate="2000-01-01", edate="2010-12-31") %>% as_tibble()
plot(sflow$dates, sflow$val, type='l')
```
